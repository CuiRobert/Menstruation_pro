```python
import torch
import torch.nn.functional as Fun
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import KFold
```

```python
from collections import Counter
```

```python
def pre_data(path1,img_prd_path):
    df = pd.read_excel(path1)
#     df.drop_duplicates(subset=['住院号'], keep='first', inplace=True)
    df2 = pd.read_excel(img_prd_path)
#     df = pd.merge(df,df2,on=['住院号','对应编号','label'],how='left')
    df = pd.merge(df,df2,how='left')
    print("数据长度：",len(df))
#     df.to_excel('../data/输入网络测试数据.xlsx',index=False)
    df_x = df[['t1','t2','p1','p2','时间差','predprob_0','predprob_1']]
#     print(df.columns.values)
    df_y = df[['label']]
    X = np.array(df_x)
    Y = np.array(df_y)
    X_tensor = torch.tensor(np.array(X),dtype=torch.float32)
    Y_tensor = torch.tensor(np.array(Y)).squeeze(dim=1)
    return X_tensor, Y_tensor
```

```python
def get_acc(outputs, labels):
    """计算acc"""
    _,predict = torch.max(outputs.data,1)
    total_num = labels.shape[0]*1.0
    correct_num = (labels == predict).sum().item()
    acc = correct_num / total_num
    label_0,label_1 = 0,0
    cre_0,cre_1 = 0,0
    for i,j in zip(predict,labels):
        if j == 0:
            label_0 += 1
            if i == j:
                cre_0 += 1
        if j == 1:
            label_1 += 1
            if i == j:
                cre_1 += 1
#     print('结果长度',len(predict.numpy()))
#     pd.DataFrame({'data_pred':list(predict.numpy().tolist()),
#                   'labels':list(labels.numpy().tolist())}).to_excel('../result/数据-图像-推算结果.xlsx',index=False)
#     df_save = pd.read_excel('../data/输入网络测试数据.xlsx')
#     df_save['data_pred'] = list(predict.numpy().tolist())
#     df_save.to_excel('../data/数据推算预测结果.xlsx',index=False)
    print(acc)
    print('总的准确率:%.4f\t (%d + %d)\t/%d'%((cre_0 + cre_1) / len(predict), cre_0, cre_1, len(predict)))
    print('标签为0的准确率：%.4f\t(%d/%d)'%((cre_0 / label_0),cre_0,label_0))
    print('标签为1的准确率：%.4f\t(%d/%d)'%((cre_1 / label_1),cre_1,label_1))
    return acc

#     return acc
```

```python
def get_acc_test(outputs, labels):
    """计算acc"""
    _,predict = torch.max(outputs.data,1)
    total_num = labels.shape[0]*1.0
    correct_num = (labels == predict).sum().item()
    acc = correct_num / total_num
    label_0,label_1 = 0,0
    cre_0,cre_1 = 0,0
    for i,j in zip(predict,labels):
        if j == 0:
            label_0 += 1
            if i == j:
                cre_0 += 1
        if j == 1:
            label_1 += 1
            if i == j:
                cre_1 += 1
    df1 = pd.DataFrame({'data_pred':list(predict.numpy().tolist()),
                          'labels':list(labels.numpy().tolist())})
    df1.to_excel('../result/数据-图像-推算结果.xlsx',index=False)
#     pd.DataFrame({'data_pred':list(predict.numpy().tolist()),
#                   'labels':list(labels.numpy().tolist())}).to_excel('../result/时间推算结果.xlsx',index=False)
#     df1.to_excel('../data/时间推算结果.xlsx',index=False)
#     df_save = pd.read_excel('../data/输入网络测试数据.xlsx')
#     df_save['data_pred'] = list(df1['data_pred'])
#     df_save.to_excel('../data/数据推算预测结果.xlsx',index=False)
    print(acc)
    print('总的准确率:%.4f\t (%d + %d)\t/%d'%((cre_0 + cre_1) / len(predict), cre_0, cre_1, len(predict)))
    print('标签为0的准确率：%.4f\t(%d/%d)'%((cre_0 / label_0),cre_0,label_0))
    print('标签为1的准确率：%.4f\t(%d/%d)'%((cre_1 / label_1),cre_1,label_1))
    return acc
```

```python
def result_combat(QT_ZQ_df, keyword='增'):

    """按病例划分结果"""
    """keyword 表示预测为1 的文件 例如‘绝’，或者'增' """
    # org_df = pd.read_excel('./result_data/分泌期_增生期_二分类_数据对应表.xlsx')
    # QT_df = pd.read_excel(QT_path)
    # ZQ_df = pd.read_excel(ZQ_path)
    #
    # QT_df = pd.merge(QT_df, org_df, how='left')
    # ZQ_df = pd.merge(ZQ_df, org_df, how='left')
    QT_ZQ_df = QT_ZQ_df

    case2img = {}  # 存储病例编号，及对应图片
    for i, row in QT_ZQ_df.iterrows():
        case_id = row['对应编号']
        if case_id in case2img:
            case2img[case_id].append(row['newname'])
        else:
            case2img[case_id] = [row['newname']]

    case2pred = {}  # 存储病例编号，及对应图片预测值
    for key, value in case2img.items():
        temp_label = []
        for i in case2img[key]:
            temp_label.append(QT_ZQ_df[QT_ZQ_df['newname'] == i]['com_pred'].values[0])
        case2pred[key] = temp_label

    case2result = {}  # 存储病例编号，及最终结果
    for key, value in case2pred.items():
        countDict = Counter(sorted(value, reverse=True))
        case2result[key] = max(countDict, key=lambda x: countDict[x])


    correct_JJ = 0
    correst_QT = 0
    count_label_1 = 0
    for _, value in case2result.items():
        if keyword in _:
            count_label_1 += 1
        if keyword in _ and value == 1:
            correct_JJ += 1
        if keyword not in _ and value == 0:
            correst_QT += 1

    print('总的准确率：%.4f\t (%d + %d)\t/%d ' % (
    (correct_JJ + correst_QT) / len(case2result), correct_JJ, correst_QT, len(case2result)))
    
    print('标签为0的准确率：%.4f\t(%d/%d)' % (
    (correst_QT / (len(case2result) - count_label_1)), correst_QT, (len(case2result) - count_label_1)))
    print('标签为1的准确率(绝经期或者增生期)：%.4f\t(%d/%d)' % (correct_JJ / count_label_1, correct_JJ, count_label_1))
```

```python
def combat_camp_reslut(outputs,testpath_img_prod):
    _,predict = torch.max(outputs.data,1)
    df_te = pd.read_excel(testpath_img_prod)
    df_te['com_pred'] = list(predict)
    result_combat(df_te,'增')
```

```python
# 定义BP神经网络
class Net(torch.nn.Module):
    def __init__(self, n_feature, n_hidden, n_output):
        super(Net, self).__init__()
        self.hidden = torch.nn.Linear(n_feature, n_hidden)
        self.out = torch.nn.Linear(n_hidden, n_output)
        
    def forward(self,x):
        x = Fun.relu(self.hidden(x))
        x = self.out(x)
        return x
```

```python
def main():
    trainpath_data = '../data/图像输出结果/训练集——月经周期推算（唯一编号）一一对应.xlsx'
    trainpath_img_prod ='../data/图像输出结果/训练集概率值.xlsx'
    testpath_data = '../data/图像输出结果/测试集——月经周期推算（唯一编号）一一对应.xlsx'
    testpath_img_prod ='../data/图像输出结果/测试集概率值.xlsx'
    
    X_trian,Y_trian = pre_data(trainpath_img_prod,trainpath_data)
    X_test,Y_test = pre_data(testpath_img_prod,testpath_data)
    
    
    acc_total = 0
    epoch = 3
    
    kf = KFold(n_splits=epoch,shuffle=False,random_state=33)
    for train_index,test_index in kf.split(X_trian):
        x1,y1 = X_trian[train_index],Y_trian[train_index]
        x2,y2 = X_trian[test_index],Y_trian[test_index]
#         x1,x2,y1,y2 = train_test_split(X_trian,Y_trian,test_size=0.33)
        net = Net(n_feature=7,n_hidden=20,n_output=2)
        optimizer = torch.optim.SGD(net.parameters(), lr = 0.05)
        loss_func = torch.nn.CrossEntropyLoss()
        
        for i in range(1000):
            out = net(x1)
            loss = loss_func(out, y1)
            # 输出与label对比
            optimizer.zero_grad()
            # 初始化
            loss.backward()
            optimizer.step()

        out = net(x2)
        acc_1 = get_acc(out,y2)
        acc_total += acc_1
        print('_'*10)
        
    print('交叉验证%.4f'%(acc_total/epoch))
    print('_'*20)
    out = net(X_test)
    torch.save(net.state_dict(),'../models/' + 'test111.pkl')
    get_acc_test(out,Y_test)
    
    
    net2 = Net(n_feature=7,n_hidden=20,n_output=2)
    net2.load_state_dict(torch.load('../models/' + 'test111.pkl'))
    out2 = net(X_test)
    get_acc_test(out2,Y_test)
    print('___'*10)
    combat_camp_reslut(out2,testpath_img_prod)
```

```python
def get_prob_test(outputs, labels):
    """计算acc"""
    _,predict = torch.max(outputs.data,1)
    total_num = labels.shape[0]*1.0
    correct_num = (labels == predict).sum().item()
    acc = correct_num / total_num
    label_0,label_1 = 0,0
    cre_0,cre_1 = 0,0
    for i,j in zip(predict,labels):
        if j == 0:
            label_0 += 1
            if i == j:
                cre_0 += 1
        if j == 1:
            label_1 += 1
            if i == j:
                cre_1 += 1
    df1 = pd.DataFrame({'data_pred':list(predict.numpy().tolist()),
                          'labels':list(labels.numpy().tolist())})
    df1['prob_0'] = list(i.item() for i in outputs.data[:,0])
    df1['prob_1'] = list(i.item() for i in outputs.data[:,1])
#     df1.to_excel('../result/分泌增生-数据-图像-概率结果.xlsx',index=False)
    
    testpath_img_prod ='../data/图像输出结果/测试集概率值.xlsx'
    df_save = pd.read_excel(testpath_img_prod)
    df_save['labels'] = df1['labels']
    df_save['prob_0'] = df1['prob_0']
    df_save['prob_1'] = df1['prob_1']
    df_save_ZQ = df_save[df_save['orgname'].str.contains('纵')]
    df_save_QT = df_save[~df_save['orgname'].str.contains('纵')]
    
    df_save.to_excel('../result/分泌增生-数据-图像-概率结果.xlsx',index=False)
    df_save_ZQ.to_excel('../result/分泌增生-数据-图像-概率结果-纵切.xlsx',index=False)
    df_save_QT.to_excel('../result/分泌增生-数据-图像-概率结果-其他.xlsx',index=False)
#     pd.DataFrame({'data_pred':list(predict.numpy().tolist()),
#                   'labels':list(labels.numpy().tolist())}).to_excel('../result/时间推算结果.xlsx',index=False)
#     df1.to_excel('../data/时间推算结果.xlsx',index=False)
#     df_save = pd.read_excel('../data/输入网络测试数据.xlsx')
#     df_save['data_pred'] = list(df1['data_pred'])
#     df_save.to_excel('../data/数据推算预测结果.xlsx',index=False)
    print(acc)
    print('总的准确率:%.4f\t (%d + %d)\t/%d'%((cre_0 + cre_1) / len(predict), cre_0, cre_1, len(predict)))
    print('标签为0的准确率：%.4f\t(%d/%d)'%((cre_0 / label_0),cre_0,label_0))
    print('标签为1的准确率：%.4f\t(%d/%d)'%((cre_1 / label_1),cre_1,label_1))
    return acc
```

```python
def test_main():
    testpath_data = '../data/图像输出结果/测试集——月经周期推算（唯一编号）一一对应.xlsx'
    testpath_img_prod ='../data/图像输出结果/测试集概率值.xlsx'
    X_test,Y_test = pre_data(testpath_img_prod,testpath_data)
    net2 = Net(n_feature=7,n_hidden=20,n_output=2)
    net2.load_state_dict(torch.load('../models/' + 'test111.pkl'))
    out2 = net2(X_test)
    get_prob_test(out2,Y_test)
    print('___'*10)
    combat_camp_reslut(out2,testpath_img_prod)
    
```
