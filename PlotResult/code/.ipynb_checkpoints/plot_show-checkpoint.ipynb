```python
import pandas as pd
import numpy as np
import os
import matplotlib
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import roc_curve, auc,roc_auc_score
```

```python
def softmax2prob(path_df):
    df = pd.read_excel(path_df)
    for i, row in df.iterrows():
        df.loc[i,'prob_0'] = np.exp(row['predprob_0'])/(np.exp(row['predprob_0']) + np.exp(row['predprob_1']))
        df.loc[i,'prob_1'] = np.exp(row['predprob_1'])/(np.exp(row['predprob_0']) + np.exp(row['predprob_1']))
    df.to_excel('../result/图像输出概率值.xlsx',index = False)
```

```python
path1 = '../data/测试集概率值.xlsx'
softmax2prob(path1)
```

```python
def plt_pr(path_df):
    df = pd.read_excel(path_df)
    plt.figure("P-R Curve")
    plt.title('Precision/Recall Curve')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    
    y_true = np.array(df['label'])
    y_scores = np.array(df['prob_1'])
    precision, recall, thresholds = precision_recall_curve(y_true, y_scores)
    plt.plot(recall,precision)
    plt.show()
```

```python
def plt_auc(path_df):
    df = pd.read_excel(path_df)
    y_label = np.array(df['label'])
    y_pre = np.array(df['prob_1'])
    
    
    fpr, tpr, thersholds = roc_curve(y_label, y_pre)
 
#     for i, value in enumerate(thersholds):
#         print("%f %f %f" % (fpr[i], tpr[i], value))

    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, label='ROC (area = {0:.4f})'.format(roc_auc), lw=2)

    plt.xlim([-0.05, 1.05])  # 设置x、y轴的上下限，以免和边缘重合，更好的观察图像的整体
    plt.ylim([-0.05, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')  # 可以使用中文，但需要导入一些库即字体
    plt.title('ROC Curve')
    plt.legend(loc="lower right")
    plt.show()
```

```python
def roc_auc_ci(y_true, y_score, positive=1):
    AUC = roc_auc_score(y_true, y_score)
    N1 = sum(y_true == positive)
    N2 = sum(y_true != positive)
    Q1 = AUC / (2 - AUC)
    Q2 = 2*AUC**2 / (1 + AUC)
    SE_AUC = np.sqrt((AUC*(1 - AUC) + (N1 - 1)*(Q1 - AUC**2) + (N2 - 1)*(Q2 - AUC**2)) / (N1*N2))
    lower = AUC - 1.96*SE_AUC
    upper = AUC + 1.96*SE_AUC
    if lower < 0:
        lower = 0
    if upper > 1:
        upper = 1
    return (lower, upper)
```

```python
def print_auc(path):
    df = pd.read_excel(path)
    try:
        y_label, y_pre= np.array(df['label']), np.array(df['probablity1'])
    except:
        y_label, y_pre= np.array(df['label']), np.array(df['prob_1'])
    (lower,upper) = roc_auc_ci(y_label,y_pre)
    print("%.4f-%.4f"%(lower,upper))
```
